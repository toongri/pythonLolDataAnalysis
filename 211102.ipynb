```python
url = "https://api.upbit.com/v1/candles/minutes/1"

querystring = {"market":"KRW-ETH","count":"200"}

headers = {"Accept": "application/json"}

response = requests.request("GET", url, headers=headers, params=querystring)
```

- coin_name Î¶¨Ïä§Ìä∏ÏóêÏÑú forÎ¨∏ Ïù¥Ïö© ÌïòÎÇòÏî© Í∫ºÎÇ¥ÏôÄÏÑú querystring marketÏùò vlaue('krw-eth')Ïóê ÎåÄÏûÖ
- responseÎ•º Î∞õÍ≥† .json() Îç∞Ïù¥ÌÑ∞Î•º Î≥ÄÌôò

- df1.append(df2)

```python
df = pd.DataFrame()
for c in coin_name:
    url = "https://api.upbit.com/v1/candles/minutes/1"
    querystring = {"market":c,"count":"200"}
    headers = {"Accept": "application/json"}
    response = requests.request("GET", url, headers=headers, params=querystring)
    tmp_df = pd.DataFrame(response.json())
    df = df.append(tmp_df)
```

```python
df
```

```python
list(df.columns)
```

```python
my_df = df[['market','candle_date_time_kst','opening_price','high_price','low_price','trade_price']].copy()
my_df
```

- candle_date_time_kst
- '2021-11-03T09:17:00' -> '09:17:00'
- forÎ¨∏ÏùÄ ÏÜçÎèÑÍ∞Ä ÎäêÎ¶º
- Ïª¨Îüº Ï†ÑÏ≤¥ ÏÑ†ÌÉùÌï¥ÏÑú ÏùºÍ¥ÑÏûëÏóÖ
- replace('2021-11-03T','')
- Ïñ¥Ï†ú ÎàÑÏ†ÅÌôïÏßÑÏûê ex) '2,300' -> 2000

```python
for i in range(len(my_df)):
    st = my_df.iloc[i].candle_date_time_kst
    st2 = st.split('T')[-1]
    my_df['candle_date_time_kst'].iloc[i] = st2
```
