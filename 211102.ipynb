- 2000 ~ 8000
- ~3000 = 주의
- 3000~5000 = 매우 주의
- 5000~6000 = 높음
- 6000~ = 매우 높음

```python
def calc_danger(x):
    if x > 6000:
        return '매우 높음'
    elif x > 5000:
        return '높음'
    elif x > 3000:
        return '매우 주의'
    else:
        return '주의'
```

```python
df['위험도'] = df['확진자 수'].apply(lambda x: calc_danger(x))
```

```python
coin_name = ['KRW-ETH','KRW-BTC','KRW-XRP']
```

```python
url = "https://api.upbit.com/v1/candles/minutes/1"

querystring = {"market":"KRW-ETH","count":"200"}

headers = {"Accept": "application/json"}

response = requests.request("GET", url, headers=headers, params=querystring)
```

- coin_name 리스트에서 for문 이용 하나씩 꺼내와서 querystring market의 vlaue('krw-eth')에 대입
- response를 받고 .json() 데이터를 변환
- dataframe으로 변환
- df1.append(df2)
- df = pd.DataFrame()

```python
df = pd.DataFrame()
for c in coin_name:
    url = "https://api.upbit.com/v1/candles/minutes/1"
    querystring = {"market":c,"count":"200"}
    headers = {"Accept": "application/json"}
    response = requests.request("GET", url, headers=headers, params=querystring)
    tmp_df = pd.DataFrame(response.json())
    df = df.append(tmp_df)
```

```python
df
```

```python
list(df.columns)
```

```python
my_df = df[['market','candle_date_time_kst','opening_price','high_price','low_price','trade_price']].copy()
my_df
```

- candle_date_time_kst
- '2021-11-03T09:17:00' -> '09:17:00'
- for문은 속도가 느림
- 컬럼 전체 선택해서 일괄작업
- replace('2021-11-03T','')
- 어제 누적확진자 ex) '2,300' -> 2000

```python
for i in range(len(my_df)):
    st = my_df.iloc[i].candle_date_time_kst
    st2 = st.split('T')[-1]
    my_df['candle_date_time_kst'].iloc[i] = st2
```

```python
my_df
```

```python
my_df
```

```python
my_df['candle_date_time_kst'] = my_df.candle_date_time_kst.str.replace('2021-11-03T','')
```

```python
my_df
```

```python
my_df['candle_date_time_kst'] = my_df['candle_date_time_kst'].str[11:]
```
