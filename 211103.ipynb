- 전월세가 전세인것만 df 분리
- 분리된 df 법정동명 기준으로 groupby 보증금 평균
- sort 보증금 높은 순 정렬

```python
tmp = my_df[my_df['전월세']=='전세']
tmp
```

```python
tmp.groupby(['법정동명'])['보증금'].mean()
```

```python
type(tmp['보증금'].iloc[0])
```

```python
tmp = tmp.astype({'보증금':'int'})
```

```python
type(tmp['보증금'].iloc[0])
```

```python
group_df = tmp.groupby(['법정동명'])[['보증금']].mean().round(2)
group_df
```

```python
group_df.sort_values(by=['보증금'], ascending=False)
```
