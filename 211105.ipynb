```python
res['전국대비서울확진비율'] = round(res['서울시확진자']/res['전국확진'], 2)
res
```

```python
res['전국대비서울사망률'] = round(res['서울시사망']/res['전국사망'], 2)
res['서울시사망률'] = round(res['서울시사망']/res['서울시확진자'], 2)
res['서울시퇴원율'] = round(res['서울시퇴원']/res['서울시확진자'], 2)
```
